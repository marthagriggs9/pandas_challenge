```python
#Dependencies and Setup
import pandas as pd
```

```python
#File to load
#Save path to data set in a variable
#School data
school_data_to_load = 'Resources/schools_complete.csv'
#Student data
student_data_to_load = 'Resources/students_complete.csv'
```

```python
#Read the school data file and store it in a Pandas DataFrame
school_data_df = pd.read_csv(school_data_to_load)
school_data_df
```

```python
#Read the student data file and store it in a Pandas DataFrame
student_data_df = pd.read_csv(student_data_to_load)
student_data_df.head()
```

```python
#Check to see if there are any missing values in the school data
school_data_df.count()
```

```python
#Check to see if there are any missing values in the student data
student_data_df.count()
```

```python
#Merge the data into a single dataset
school_data_merged = pd.merge(student_data_df, school_data_df, how="left", 
                              on=["school_name", "school_name"])
school_data_merged.head()
```

## District Summary

```python
#Calculate the total number of unique schools
school_count = len(school_data_merged["school_name"].unique())
school_count
```

```python
#Calculate the total number of students
student_count = school_data_merged["student_name"].count()
student_count
```

```python
#Calculate the total budget
#Only look at the original school data file because 
#that lists the budget for each school.
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#Find the average math score (mean)
average_math_score = school_data_merged["math_score"].mean()
average_math_score
```

```python
#Find the average reading score (mean)
average_reading_score = school_data_merged["reading_score"].mean()
average_reading_score
```

```python
#Find the percentage of students who passed math (scored greater than or equal to 70)
passing_math = school_data_merged[(school_data_merged["math_score"] >= 70)].count()["student_name"]
passing_math
passing_math_percentage = passing_math / float(student_count) * 100
passing_math_percentage
```

```python
#Find the percentage of students who passed reading (scored greater than or equal to 70)
passing_reading = school_data_merged[(school_data_merged["reading_score"] >= 70)].count()["student_name"]
passing_reading
passing_reading_percentage = passing_reading / float(student_count) * 100
passing_reading_percentage
```

```python
#Find the percentage of students who passed both math AND reading
passing_math_and_reading_count = school_data_merged[
    (school_data_merged["math_score"] >=70) & (school_data_merged["reading_score"] >= 70)
    ].count()["student_name"]
overall_passing_rate = passing_math_and_reading_count / float(student_count) * 100
overall_passing_rate
```

```python
#Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_df = pd.DataFrame(
    [{"Total Schools" : school_count, 
      "Total Students" : student_count, 
      "Total Budget" : total_budget, 
      "Average Math Score" : average_math_score, 
      "Average Reading Score" : average_reading_score, 
      "% Passing Math" : passing_math_percentage, 
      "% Passing Reading" : passing_reading_percentage, 
      "% Overall Passing" : overall_passing_rate}])

#Format the values in the new DataFrame
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)


#New Data Frame completed
district_summary_df
```

## School Summary

```python
#Determine the type of school
school_types = school_data_df.set_index(["school_name"])["type"]
school_types
```

```python
#Create a data frame that contains the school name and type information
type_df = pd.DataFrame(school_types)
type_df
```

```python
#Calculate the total school budget and per capita spending
per_school_budget = school_data_merged.groupby(["school_name"]).mean()["budget"]

per_school_budget
```

```python
#Calculate the per capita spending (per student budget)
per_school_capita = per_school_budget/students_per_school
per_school_capita
```

```python
#Find the average reading score per school
per_school_reading = school_data_merged.groupby(["school_name"]).mean()["reading_score"]
per_school_reading
```

```python
#Find the average math score per school
per_school_math = school_data_merged.groupby(["school_name"]).mean()["math_score"]
per_school_math
```

```python
#Find the number of schools with math scores of 70 or higher
school_passing_math = school_data_merged[(school_data_merged["math_score"] >= 70)]
school_passing_math
#Calculate the passing rate for students passing math
school_passing_math = school_passing_math.groupby(["school_name"]).count()["student_name"] / students_per_school *100
school_passing_math
```

```python
#Find the number of schools with reading scores of 70 or higher
school_passing_reading = school_data_merged[(school_data_merged["reading_score"] >= 70)]
school_passing_reading
#Calculate the passing rate for students passing reading
school_passing_reading = school_passing_reading.groupby(["school_name"]).count()["student_name"] / students_per_school *100
school_passing_reading
```

```python
#Find the number of schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_merged[
    (school_data_merged["reading_score"] >= 70) & (school_data_merged["math_score"] >= 70)]
passing_math_and_reading
#Calculate the passing rate for students passing both math and reading
passing_math_and_reading = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / students_per_school *100
passing_math_and_reading
```

```python
#Create a DataFrame that summarizes the key metrics about each school
per_school_summary = pd.DataFrame(
    {"School Type" : school_types,
      "Total Students" : students_per_school, 
      "Total School Budget" : per_school_budget, 
      "Per Student Budget" : per_school_capita, 
      "Average Math Score" : per_school_math, 
      "Average Reading Score" : per_school_reading, 
      "% Passing Math" : school_passing_math, 
      "% Passing Reading" : school_passing_reading, 
      "% Overall Passing" : passing_math_and_reading})

per_school_summary

```

```python
#Format the Total School Budget and Per Student Budget Columns
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)

per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary.head()
```

## Scores by School Spending

```python
#Get descriptive statistics for the per_school_capita
per_school_capita.describe()
```

```python
#Cut the per_school_capita into the spending ranges
spending_bins = [0, 585, 630, 645, 680]
pd.cut(per_school_capita, spending_bins)
```

```python
#Cut the per_school_capita into the spending ranges
spending_bins = [0, 585, 630, 645, 680]
per_school_capita.groupby(pd.cut(per_school_capita, spending_bins)).count()
```

```python
#Create a table that breaks down school performance based on average spending ranges(per student)
#Establish the bins

spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
#Use pd.cut to categorize spending based on the bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=group_names) 
school_spending_df
```

```python
#Calculate the mean scores per spending range 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
#Use the scores to create a DataFrame called "spending_summary"

spending_summary = pd.DataFrame({
    "Average Math Score" : spending_math_scores, 
    "Average Reading Score" : spending_reading_scores, 
    "% Passing Math" : spending_passing_math, 
    "% Passing Reading" : spending_passing_reading, 
    "% Overall Passing" : overall_passing_spending})

#Display results
spending_summary
```

```python
#Calculate averages for the desired columns
size_math_scores = per_school_summary.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = per_school_summary.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = per_school_summary.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = per_school_summary.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = per_school_summary.groupby(["School Size"]).mean()["% Overall Passing"]
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`

size_summary = pd.DataFrame({
    "Average Math Score" : size_math_scores, 
    "Average Reading Score" : size_reading_scores, 
    "% Passing Math" : size_passing_math, 
    "% Passing Reading" : size_passing_reading, 
    "% Overall Passing" : size_overall_passing})

# Display results
size_summary
```
